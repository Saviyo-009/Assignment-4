### Visualization of both the Datasets

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Neighbourhood Group', y='Price', data=airbnb_df)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Box Plot of Prices by Neighbourhood Group')
plt.grid(True)
plt.show()
```

#### The above visualization shows the distribution of Price column with Neighborhood Groups column from the Airbnb dataframe.
#### The purpose of visualization is to identify any potential differences or patterns in pricing, and understand the spread of prices within each neighbourhood group.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Minimum Nights', data=airbnb_df)
plt.xlabel('Minimum Nights')
plt.ylabel('Count')
plt.title('Count of Minimum Nights of Stay Requirement')
plt.xticks(rotation=45)
plt.show()
```

#### The above visualization displays the count of Minimum Nights column from the Airbnb dataframe.
#### The purpose of visualization is to identify common stay requirements and determine the most common duration of stays that hosts offer in the dataset.

```python
plt.figure(figsize=(20, 8))
sns.lineplot(x='Neighbourhood', y='Availability 365', data=airbnb_df, color = 'green', marker = 'o',mfc = 'c',ms=10)
plt.xlabel('Neighbourhood')
plt.ylabel('Availability 365')
plt.title('Availability of Listings by Neighbourhood')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

#### The above visualization displays the Neighbourhood column  and the  Availability 365 column from the Airbnb dataframe.
#### The purpose of visualization is to identify number of days the listings are available for booking in the next 365 days from the neighbourhood.

```python
avg_price_by_room = airbnb_df.groupby('Room Type')['Price'].mean().reset_index()
sns.barplot(x='Room Type', y='Price', data=avg_price_by_room)
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.title('Average Price by Room Type')
plt.show()
```

#### The above visualization shows the average price for the Room Type column from the Airbnb dataframe.
#### The purpose of visualization is to identify the differences in average prices among the different room types.

```python
hr_df['Date of Hire'] = pd.to_datetime(hr_df['Date of Hire'])
hires_per_date = hr_df['Date of Hire'].value_counts().sort_index()
plt.figure(figsize=(12, 5))
plt.fill_between(hires_per_date.index, hires_per_date, alpha=0.5,color='r')
plt.xlabel('Date of Hire')
plt.ylabel('Number of Hires')
plt.title('Area Chart of Number of Hires Over Time')
plt.xticks(rotation=45)
plt.show()
```

#### The above visualization represents the Date of Hire column and the Number of Hires column from the HR dataframe.
#### The purpose of visualization is to show the cumulative sum of hires over time, showcasing the overall growth or decline in hiring over the observed period.

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='Performance Score', y='Salary',hue='Sex', data=hr_df)
plt.xlabel('Performance Score')
plt.ylabel('Salary')
plt.title('Violin Plot of Salary by Gender and Performance Score')
plt.show()
```

#### The above visualization represents the Salary column, the Performance Score column and the Sex column from the HR dataframe.
#### The purpose of visualization is to show the distribution of salaries for each performance score based on gender.

```python
plt.hist(hr_df['Marital Desc'], color='skyblue')
plt.xlabel('Marital Desc')
plt.ylabel('Frequency')
plt.title('Histogram Describing the Marital Status')
plt.show()
```

#### The above visualization shows the frequency of the Marital Desc column from the HR dataframe.
#### The purpose of visualization is to represent the number of occurrences of a particular marital status in each bar.

```python
plt.figure(figsize=(10,6))
plt.scatter(airbnb_df['Latitude'], airbnb_df['Longitude'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Geographical Distribution of Airbnb Listings')
plt.grid(True)
plt.show()
```

#### The above visualization displays the Latitude and Longitude columns from the Airbnb dataframe.
#### The purpose of visualization is to show the geographical distribution of Airbnb listings based on their latitude and longitude coordinates. Each point on the plot represents the location of an Airbnb listing, providing an overview of how the listings are distributed across different geographical areas.

```python
recruitment_source_counts = hr_df['Recruitment Source'].value_counts()
plt.figure(figsize=(15,15))
plt.pie(recruitment_source_counts, labels=recruitment_source_counts.index, autopct='%1.1f%%')
plt.title('Recruitment Source Distribution')
plt.show()
```

#### The above visualization represents the value counts of the Recruitment Source column from the HR dataframe.
#### The purpose of visualization is to show the distribution of recruitment sources used to hire employees.

```python
Employee_count = hr_df.groupby(['Department', 'EmploymentStatus'])['Emp ID'].count().unstack()
ax = Employee_count.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Stacked Bar Chart of Employment Status by Department')
plt.legend(title='Employment Status')
plt.xticks(rotation=45)
plt.show()
```

#### The above visualization represents the Employment Status and Department Columns with respect to number of employees from the HR dataframe.
#### The purpose of visualization is to display the count of employees in different employment status categories i.e Active, Terminated for Cause and Voluntarily Terminated  within that department.
